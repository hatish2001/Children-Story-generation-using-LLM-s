```python
"""
  Wordcloud function for the datasets

"""
from wordcloud import WordCloud
def wordcloud_function1(file_path, title, fontsize=15, figsize=(10, 10), width=800, height=400, max_words=200, max_font_size=100):
    """
      This function opens and reads a text file, extracts the text, and generates a word cloud.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    wordcloud = WordCloud(width=width, height=height).generate(text)

    plt.figure(figsize=figsize)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=fontsize)
    plt.axis("off")
    plt.show()

file_path1 = '/content/nlpProject 3.txt'
wordcloud_function1(file_path1, title="Wordcloud of Our Preprocessed Data")
file_path2 = 'cleaned_merged_fairy_tales_without_eos.txt'
wordcloud_function1(file_path2, title="Children Stories Text from Project Gutenberg")
```

```python
"""
Analysis of the surveys

"""

import matplotlib.pyplot as plt
import numpy as np

categories = ['Grammatically Correct', 'Flows Well', 'Kid-Friendly']
models = ['text-davinci-003', 'babbage', 'human written']
surveys = [
    [[6, 0, 4], [7, 0, 3], [0, 0, 10]],
    [[6, 1, 3], [6, 0, 4], [6, 0, 4]],
    [[1, 3, 6], [1, 1, 8], [2, 1, 7]]
]

surveys_array = np.array(surveys)

fig, axs = plt.subplots(nrows=len(surveys), ncols=1, figsize=(12, 8 * len(surveys)))

bar_width = 0.2
bar_positions = np.arange(len(categories))

for i, survey in enumerate(surveys_array):
    for j, model in enumerate(models):
        axs[i].bar(bar_positions + j * bar_width, survey[:, j], bar_width, label=f'{model}')

    axs[i].set_xlabel('Categories', fontdict={'weight': 'bold', 'style': 'italic', 'size': 14})
    axs[i].set_ylabel('Count', fontdict={'weight': 'bold', 'style': 'italic', 'size': 14})
    axs[i].set_title(f'Model and Human Text Ratings - Survey {i + 1}', fontdict={'weight': 'bold', 'style': 'italic', 'size': 16})
    axs[i].set_xticks(bar_positions + ((len(models) - 1) * bar_width) / 2)
    axs[i].set_xticklabels(categories, fontsize=14)
    axs[i].set_yticklabels(axs[i].get_yticks(), fontsize=14)
    axs[i].legend(fontsize='large')

plt.tight_layout()

plt.show()
```
