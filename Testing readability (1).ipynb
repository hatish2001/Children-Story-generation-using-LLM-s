## Text Preprocessing Steps

- Removing non-alphabetic characters
- Converting text to lowercase
- Tokenizing text
- Removing stopwords
- Lemmatizing words
- Joining the processed words back together


```python
def custom_preprocessing(df):
    """
    This function processes the text data in the given dataframe, making it suitable for machine learning tasks.
    It includes steps like removing non-alphabetic characters, converting to lowercase, tokenization,
    removing stopwords, and lemmatization.
    """

    processed_texts = []
    stop_words = set(stopwords.words("english"))
    lemmatizer = nltk.WordNetLemmatizer()

    for text in tqdm(df['excerpt'].values):
        # Remove non-alphabetic characters
        text = re.sub('[^a-zA-Z]', ' ', text)

        # Convert to lowercase
        text = text.lower()

        # Tokenization
        words = nltk.word_tokenize(text)

        # Remove stopwords
        words = [word for word in words if word not in stop_words]

        # Lemmatization
        words = [lemmatizer.lemmatize(word) for word in words]

        # Join the processed words
        processed_text = " ".join(words)
        processed_texts.append(processed_text)

    processed_texts_series = pd.Series(processed_texts, name='Converted_text')
    return processed_texts_series
```

## Feature Extraction
- Counting occurrences of punctuation and quotes
- Extracting the length of the longest word
- Creating a new dataframe with extracted features

```python
def extract_text_features(input_df, stop_words):
    """
    Takes a DataFrame and a set of stop words, processes the 'excerpt' column,
    and extracts various features such as the number of sentences, words,
    lemma length, and overall preprocessed essay length.
    """
    sentence_lengths = []
    num_words_list = []
    sentence_length_list = []
    word_length_list = []
    lemma_length_list = []
    num_lemmas_list = []
    preprocessed_essay_lengths = []
    initial_text_lengths = []
    num_sentences_list = []
    text_shortage_list = []

    for original_text in tqdm(input_df['excerpt'].values):

        initial_length = len(original_text)
        initial_text_lengths.append(initial_length)
        num_sentences = len(sent_tokenize(original_text))
        num_sentences_list.append(num_sentences)

        modified_text = re.sub('[^a-zA-Z]', ' ', original_text)
        modified_text = modified_text.lower()
        modified_text = word_tokenize(modified_text)

        num_words = len(modified_text)
        num_words_list.append(num_words)
        sentence_length_list.append(num_words/num_sentences)
        word_length_list.append(initial_length/num_words)

        modified_text = [word for word in modified_text if not word in stop_words]
        lemmatizer = nltk.WordNetLemmatizer()
        modified_text = [lemmatizer.lemmatize(word) for word in modified_text]

        num_lemmas = len(modified_text)
        num_lemmas_list.append(num_lemmas)
        modified_text = " ".join(modified_text)

        preprocessed_essay_length = len(modified_text)
        preprocessed_essay_lengths.append(preprocessed_essay_length)

        lemma_length_list.append(preprocessed_essay_length/num_lemmas)
        text_shortage_list.append(preprocessed_essay_length/initial_length)

    final_features_df = pd.concat([pd.Series(sentence_length_list), pd.Series(num_words_list),
                             pd.Series(word_length_list), pd.Series(lemma_length_list),
                             pd.Series(num_sentences_list), pd.Series(initial_text_lengths),
                             pd.Series(num_lemmas_list), pd.Series(preprocessed_essay_lengths),
                             pd.Series(text_shortage_list)], axis=1)
    final_features_df.columns = ["sentence_length", "num_of_words", "word_length",
                                 "lemma_length", "num_of_sentences",
                                 "initial_text_length", "num_of_lemmas",
                                 "preprocessed_essay_length", "text_shortage"]

    return final_features_df
```

```python
def extract_additional_features(df: pd.DataFrame):
    """
    Extracts various features from the text data in the dataframe for machine learning predictions.
    """


    # This function cound the number of punctutation and quoutes to make the machine learning model learn better
    comma_counts = []
    semicolon_counts = []
    exclamation_counts = []
    question_counts = []
    quote_counts = []
    period_counts = []
    longest_words = []

    for i in range(len(df)):
        text = df['excerpt'].iloc[i]

        # Count occurrences of punctuation and quotes
        comma_counts.append(text.count(","))
        semicolon_counts.append(text.count(";"))
        exclamation_counts.append(text.count("!"))
        question_counts.append(text.count("?"))
        quote_counts.append(text.count('"'))
        period_counts.append(text.count('.'))

        # Extract length of the longest word
        word_lengths = [len(w) for w in text.split(" ")]
        longest_words.append(np.max(word_lengths))

    # Create a new dataframe with extracted features
    features_df = pd.DataFrame({
        "num_of_commas": comma_counts,
        "num_of_semicolons": semicolon_counts,
        "num_of_exclamations": exclamation_counts,
        "num_of_questions": question_counts,
        "num_of_quotes": quote_counts,
        "num_of_periods": period_counts,
        "longest_word": longest_words
    })

    return features_df
```

```python
def expand_contractions(phrase):
    """
    Expands contractions in the given phrase for better readability and understanding.
    """
    # Specific contractions
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can't", "can not", phrase)

    # General contractions
    phrase = re.sub(r"n't", " not", phrase)
    phrase = re.sub(r"'re", " are", phrase)
    phrase = re.sub(r"'s", " is", phrase)
    phrase = re.sub(r"'d", " would", phrase)
    phrase = re.sub(r"'ll", " will", phrase)
    phrase = re.sub(r"'t", " not", phrase)
    phrase = re.sub(r"'ve", " have", phrase)
    phrase = re.sub(r"'m", " am", phrase)

    return phrase
```

## Further pre-processing 
- Removing non-alphabetic characters
- Converting text to lowercase and removing stopwords
- Lemmatization of words


```python
def custom_preprocessing1(df):
    """
    Performs basic preprocessing on the text data in the dataframe before using different vectorizers.
    This ensures optimal output values for further processing.
    """

    stop_words = set(stopwords.words('english'))
    preprocessed_text_list = []

    for i in tqdm(range(len(df['excerpt']))):
        text = df['excerpt'].iloc[i]

        # Expand contractions
        text = expand_contractions(text)

        # Remove non-alphabetic characters
        text = re.sub('[^a-zA-Z]', ' ', text)

        # Convert to lowercase and remove stopwords
        text = [word for word in text.lower().split(' ') if word not in stop_words]

        # Lemmatization
        lemmatizer = nltk.WordNetLemmatizer()
        text = [lemmatizer.lemmatize(word) for word in text]

        preprocessed_text_list.append(' '.join(text))

    return preprocessed_text_list
```

## Data Transformation
- Instructions on fitting and transforming the training set, and transforming the test set using the same scaler.

```python
def scale_features(df_train, df_test):
    """
    Scales the features in both the training and test sets to lie between 0 and 1.
    """
    scaler = MinMaxScaler()

    # Fit and transform the training set
    df_scaled_train = scaler.fit_transform(df_train)

    # Transform the test set using the same scaler
    df_scaled_test = scaler.transform(df_test)

    return df_scaled_train, df_scaled_test
```

Instructions on fitting and transforming the training set, and transforming the test set using the same vectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer

def vectorize_text(df_train, df_test):
    """
    Converts text data in both the training and test sets to bag-of-words representation using Count Vectorization.
    """
    vectorizer = CountVectorizer()

    # Fit and transform the training set
    df_vectorized_train = vectorizer.fit_transform(df_train)

    # Transform the test set using the same vectorizer
    df_vectorized_test = vectorizer.transform(df_test)

    return df_vectorized_train, df_vectorized_test
```

## Model architecture
- Training the CNN on the pre-processed data 

```python
def neural_network(X_train, y_train, Validation_data = None, metrics = ['mean_squared_error', 'mean_absolute_error'],
                   activation = 'relu', input_shape = (22561, ), optimizer = 'adam', loss = 'mean_squared_error',
                  epochs = 10, batch_size = 64, verbose = 1):
    """
    We are defining a neural network function that takes into account a different set of parameters
    that are needed to build the machine learning model and we are also giving different values
    and it would be working with different parameters and we are able to give those values to our
    deep learning models and we are going to return the output given by the model respectively.
    """

    model = Sequential()
    model.add(Dense(500, activation = activation, input_shape = input_shape))
    model.add(Dense(100, activation = activation))
    model.add(Dense(50, activation = activation))
    model.add(Dense(10, activation = activation))
    model.add(Dense(5, activation = activation))
    model.add(Dense(1))
    model.compile(loss = loss, metrics = metrics, optimizer = optimizer)
    model.fit(x = X_train, y = y_train, validation_data = Validation_data, epochs = epochs, batch_size = batch_size, verbose = verbose)
    return model
```

```python
model = neural_network(X_train = X_train, y_train = y_train, Validation_data = (X_cv, y_cv))
```
